```python
FUNCTION show_menu():
    PRINT "=== Task Manager ==="
    PRINT "1) Add task"
    PRINT "2) List tasks"
    PRINT "3) Remove task"
    PRINT "4) Save to file (optional)"
    PRINT "5) Load from file (optional)"
    PRINT "6) Pick a random task (optional)"
    PRINT "0) Exit"

FUNCTION read_int(prompt_text):
    LOOP:
        INPUT raw = input(prompt_text)
        TRY:
            choice = int(raw)
            RETURN choice
        EXCEPT ValueError:
            PRINT "Please enter a whole number."

FUNCTION add_task(tasks, next_id):
    INPUT text = input("Enter task: ").strip()
    IF text == "":
        PRINT "Task cannot be empty."
        RETURN next_id          # no change
    NEW task = {"id": next_id, "text": text}
    APPEND task TO tasks
    PRINT "Added task with ID", next_id
    RETURN next_id + 1          # increment id

FUNCTION list_tasks(tasks):
    IF LENGTH(tasks) == 0:
        PRINT "No tasks yet."
        RETURN
    PRINT "--- Your tasks ---"
    FOR each task IN tasks:
        PRINT "[" + task["id"] + "] " + task["text"]

FUNCTION find_task_index_by_id(tasks, target_id):
    # Returns the index in the list, or -1 if not found
    FOR i FROM 0 TO LENGTH(tasks)-1:
        IF tasks[i]["id"] == target_id:
            RETURN i
    RETURN -1

FUNCTION remove_task(tasks):
    IF LENGTH(tasks) == 0:
        PRINT "Nothing to remove."
        RETURN
    id_to_remove = read_int("Enter ID to remove: ")
    idx = find_task_index_by_id(tasks, id_to_remove)
    IF idx == -1:
        PRINT "No task with that ID."
        RETURN
    REMOVED = tasks[idx]
    DELETE tasks[idx]
    PRINT "Removed: [" + REMOVED["id"] + "] " + REMOVED["text"]

FUNCTION save_tasks_to_file(tasks, filename):
    # Very simple line-by-line format: "id|text"
    TRY:
        OPEN file with name filename FOR writing AS f
        FOR each task IN tasks:
            WRITE (task["id"] + "|" + task["text"] + "\n") TO f
        CLOSE f
        PRINT "Saved", LENGTH(tasks), "tasks to", filename
    EXCEPT any file error:
        PRINT "Could not save to file."

FUNCTION load_tasks_from_file(filename):
    # Returns (new_tasks_list, next_id_value)
    NEW new_tasks = []
    max_id_seen = 0
    TRY:
        OPEN file with name filename FOR reading AS f
        FOR each line IN f:
            STRIP line
            IF line == "":
                CONTINUE
            # Expect "id|text"
            SPLIT line by "|" INTO parts
            IF LENGTH(parts) < 2:
                CONTINUE      # skip malformed line
            TRY:
                parsed_id = int(parts[0])
            EXCEPT ValueError:
                CONTINUE      # skip malformed id
            task_text = parts[1]
            APPEND {"id": parsed_id, "text": task_text} TO new_tasks
            IF parsed_id > max_id_seen:
                max_id_seen = parsed_id
        CLOSE f
        next_id = max_id_seen + 1
        PRINT "Loaded", LENGTH(new_tasks), "tasks from", filename
        RETURN (new_tasks, next_id)
    EXCEPT any file error:
        PRINT "Could not load file. Starting with empty list."
        RETURN ([], 1)

FUNCTION pick_random_task(tasks):
    IF LENGTH(tasks) == 0:
        PRINT "No tasks to pick from."
        RETURN
    # Intro-level random selection approach:
    # generate a random index from 0 to len-1 using a simple helper if available.
    # If random library is allowed by your Week 10 scope, use it; otherwise simulate.
    # Pseudocode assuming 'random_index(0, n_minus_1)' is available:
    idx = random_index(0, LENGTH(tasks)-1)
    chosen = tasks[idx]
    PRINT "Try this next: [" + chosen["id"] + "] " + chosen["text"]
```

```python
# task_manager.py  (beginner-friendly)

# In-memory data
tasks = []             # list of {"id": int, "text": str}
next_id = 1            # simple increasing id

# -------- Helpers --------
def trim(text):
    return text.strip()

def normalize(text):
    return trim(text).lower()

def has_duplicate_text(candidate_text):
    cand = normalize(candidate_text)
    for item in tasks:
        if normalize(item["text"]) == cand:
            return True
    return False

def find_task_index_by_id(target_id):
    # linear search, beginner-friendly
    for i in range(len(tasks)):
        if tasks[i]["id"] == target_id:
            return i
    return -1

def list_tasks():
    if not tasks:
        print("No tasks available.")
    else:
        print("Your Tasks (" + str(len(tasks)) + "):")
        # show 1-based numbers for users
        for i in range(len(tasks)):
            item = tasks[i]
            print(str(i + 1) + ". [" + str(item["id"]) + "] " + item["text"])

# -------- Core actions --------
def add_task(text):
    global next_id

    text = trim(text)
    if text == "":
        print("Task title cannot be empty.")
        return False

    if has_duplicate_text(text):
        print("Duplicate task title. Not added.")
        return False

    tasks.append({"id": next_id, "text": text})
    print('Added: "' + text + '" (id=' + str(next_id) + ")")
    next_id += 1
    return True

def remove_task_by_number(number_str):
    """
    Users see 1-based numbering in list.
    Convert input -> int, then to 0-based index.
    """
    try:
        number = int(number_str)
    except ValueError:
        print("Please enter a whole number (e.g., 1, 2, 3).")
        return False

    # convert 1-based to 0-based
    index = number - 1

    if index < 0 or index >= len(tasks):
        if len(tasks) == 0:
            print("There are no tasks to remove.")
        else:
            print("Invalid number. Enter 1 to " + str(len(tasks)) + ".")
        return False

    removed = tasks[index]
    del tasks[index]
    print('Removed: "' + removed["text"] + '" (id=' + str(removed["id"]) + ")")
    return True

# -------- File I/O (simple, robust) --------
def save_tasks(filename):
    try:
        with open(filename, "w", encoding="utf-8") as f:
            # one task per line: id\ttext
            for item in tasks:
                line = str(item["id"]) + "\t" + item["text"] + "\n"
                f.write(line)
        print("Saved " + str(len(tasks)) + " tasks to " + filename)
        return True
    except OSError:
        print("Could not save tasks. Please check the file path or permissions.")
        return False

def load_tasks(filename):
    global tasks, next_id
    try:
        with open(filename, "r", encoding="utf-8") as f:
            # reset and rebuild
            tasks = []
            max_id = 0
            for raw in f:
                line = raw.rstrip("\n")
                if line == "":
                    continue
                parts = line.split("\t", 1)  # id, text
                if len(parts) != 2:
                    # skip malformed lines safely
                    continue
                id_part = parts[0]
                text_part = parts[1]

                # basic id check
                try:
                    tid = int(id_part)
                except ValueError:
                    # skip bad id lines
                    continue

                # avoid empty/duplicate texts when loading
                ttext = trim(text_part)
                if ttext == "" or has_duplicate_text(ttext):
                    continue

                tasks.append({"id": tid, "text": ttext})
                if tid > max_id:
                    max_id = tid

            # ensure next_id moves forward
            next_id = max_id + 1
        print("Loaded " + str(len(tasks)) + " tasks from " + filename)
        return True
    except FileNotFoundError:
        print("No existing file found‚Äîstarting with an empty list.")
        tasks = []
        next_id = 1
        return False
    except OSError:
        print("Could not read tasks. Please check the file path or permissions.")
        return False

# -------- Menu / Main Loop --------
def show_menu():
    print("\n=== Task Manager ===")
    print("1) Add task")
    print("2) List tasks")
    print("3) Remove task (by number shown in list)")
    print("4) Save tasks to file")
    print("5) Load tasks from file")
    print("6) Quit")

def main():
    while True:
        show_menu()
        choice = input("Choose an option (1-6): ").strip()

        if choice == "1":
            title = input("Enter task title: ")
            add_task(title)

        elif choice == "2":
            list_tasks()

        elif choice == "3":
            if not tasks:
                print("No tasks to remove.")
            else:
                list_tasks()
                num_str = input("Enter the task number to remove: ").strip()
                remove_task_by_number(num_str)

        elif choice == "4":
            name = input("Enter filename to save (e.g., tasks.txt): ").strip()
            if name == "":
                print("Filename cannot be empty.")
            else:
                save_tasks(name)

        elif choice == "5":
            name = input("Enter filename to load (e.g., tasks.txt): ").strip()
            if name == "":
                print("Filename cannot be empty.")
            else:
                load_tasks(name)

        elif choice == "6":
            confirm = input("Are you sure you want to quit? (y/n): ").strip().lower()
            if confirm == "y":
                print("Goodbye!")
                break
            else:
                print("Back to menu.")
        else:
            print("Invalid choice. Enter a number from 1 to 6.")

if __name__ == "__main__":
    main()
```
